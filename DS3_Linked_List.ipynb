#ADDITION OF data elements @ the END of LinkedList 

```python
# For creating a NODE we are using NODE class  
class Node:
  #Using INIT method we need to initialise the node
  def __init__(self,data): 
    # Each node contains two fields, first one is self and he second one is DATA
    self.data=data
    # None is the keyword for empty values
    self.ref=None 
    
# CREATE INDIVIDUAL NODES which connects to other nodes as a chain of nodes, for that create a LINKEDLIST class
class LinkedList:
  #Using INIT method we need to initialise the LinkedList class
  #LinkedList contains head and this head is the starting point of the LinkedList, head contains link or reference of the first node  
  def __init__(self):
    # Initially LinkedLilst will be Empty
    self.head = None

  # This method is for printing information on each node
  def print_LL(self):
    # Check if LL is empty or not
    if self.head is None:
      print("Linked List is empty")

    # If LL is not empty, then we need to go into each node and print the data inside node
    else:
      n = self.head
      while n is not None:
        print(n.data)
        n=n.ref

  def add_begin(self,data):
    #Create a node from the object of a NODE class
    new_node = Node(data)
    new_node.ref = self.head
    self.head = new_node

  def add_end(self,data):
    #Create a node from the object of a NODE class
    new_node = Node(data)
    #Before adding that to the node we need to check if the linkedlist is empty or not
    if self.head is None:
      self.head = new_node
    else:
      n = self.head
      while n.ref is not None:
        n = n.ref
      n.ref = new_node
      

LL1= LinkedList()
LL1.add_begin(10)
LL1.add_end(100)
LL1.add_begin(100)
LL1.print_LL()

```
